####    10. Постройте распределения расхода топлива по городу (city mpg) для автомобилей с ручной коробкой передач и автоматической. 
Не забудьте про подписи осей, размеры шрифтов, легенду.

```python
car[car["Transmission Type"] == "MANUAL"].hist(['city mpg'],bins=50,density=True)
plt.xlabel('city_mpg',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.title('City mpg distribution for MANUAL')
```

```python
car[car["Transmission Type"] == "AUTOMATIC"].hist(['city mpg'],bins=50,density=True)
plt.xlabel('city_mpg',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.title('City mpg distribution for AUTOMATIC')
```

```python
car[car["Transmission Type"] == "AUTOMATIC"]["city mpg"].plot.hist(alpha=0.4,density=True,bins=50,color='black',label='AUTOMATIC')
car[car["Transmission Type"] == "MANUAL"]["city mpg"].plot.hist(alpha=0.7,density=True,bins=50,color='g',label='MANUAL')
plt.legend()
plt.xlabel('city_mpg',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.title('City mpg distribution')
plt.savefig('Kotosov_fig10.png')
```

####    11. Постройте зависимость средней стоимости автомобиля от числа цилиндров. 
Используйте функцию plt.errorbar, чтобы помимо средних значений отобразить среднеквадратичное отклонение величины.

```python
#car.groupby('Engine Cylinders')['MSRP'].mean()
#lol = sorted(car['Engine Cylinders'].dropna().unique())
plt.scatter(sorted(car['Engine Cylinders'].dropna().unique()) ,car.groupby('Engine Cylinders')['MSRP'].mean()
              ,color='r')
plt.legend()
plt.xlabel('Cylinders number',fontsize=14)
plt.ylabel('Mean price',fontsize=14)
plt.errorbar(sorted(car['Engine Cylinders'].dropna().unique()), car.groupby('Engine Cylinders')['MSRP'].mean() , 
             car.groupby('Engine Cylinders')['MSRP'].std())
plt.savefig('Kotosov_fig11.png')
```

####    12. Постройте boxplot, отображающий распределение стоимости автомобилей следующих марок: Audi, Toyota, BMW, Mazda, Hyundai. 
Не забудьте про подписи осей, размеры шрифтов.

```python
price_Audi=car[(car['Make'] == 'Audi')]['MSRP'].dropna().values
price_Toyota=car[(car['Make'] == 'Toyota')]['MSRP'].dropna().values
price_BMW=car[(car['Make'] == 'BMW')]['MSRP'].dropna().values
price_Mazda=car[(car['Make'] == 'Mazda')]['MSRP'].dropna().values
price_Hyundai=car[(car['Make'] == 'Hyundai')]['MSRP'].dropna().values

fig=plt.boxplot([price_Audi,price_Toyota,price_BMW,price_Mazda,price_Hyundai],labels=['Audi','Toyota','BMW','Mazda','Hyundai'])
plt.tick_params(axis='both', which='major', labelsize=12)
plt.ylabel('Price', fontsize=14)
plt.savefig('Kotosov_fig12.png')
```
