###  2. Критерии информативности (2 балла)
На лекциях говорилось, что критерий информативности для набора объ-
ектов R вычисляется на основе того, насколько хорошо их целевые переменные предсказываются константой (при оптимальном выборе этой константы):
$$
H(R)=\frac{1}{|R|}\sum_{(x_i,y_i)\in R} L(y_i,c),
$$
где $L(y, c)$ - некоторая функция потерь и
$$
c=argmin_cH(R).
$$
 
Соответственно, чтобы получить вид критерия при конкретной функции потерь, необходимо аналитически найти оптимальное
значение константы и подставить его в формулу для H(R).
Выведите критерии информативности для следующих функций потерь:

* $L(y,c)=\sum_{k=1}^K (c_k-[y=k])^2$,

* $L(y,c)=-\sum_{k=1}^K [y=k] \log c_k$. 

* Квадратичная функция потерь $L(y,c)=\sum_{k=1}^K (c_k-[y=k])^2$

$$H(R) = \min_{\sum_k c_k = 1} \frac{1}{|R|}\sum_{(x_i, y_i) \in R}(y_i-\sum_{k=1}^Kc_k)^2$$

Очевидно, что минимум в таком выражении достигается на среденем значении целевой переменной. Соответственно, 

$$ H(R) = \frac{1}{|R|} \sum_{(x_i, y_i) \in R} \left(y_i - \frac{1}{|R|} \sum_{(x_j, y_j) \in R} y_j \right)^2 $$

$$ H(R)=\sum_{k=1}^K p_k(1-p_k) $$

* Логарифм правдоподобия $L(y,c)=-\sum_{k=1}^K [y=k] \log c_k$

$$H(R) = \min_{\sum_k c_k = 1} \left(-\frac{1}{|R|}\sum_{(x_i, y_i) \in R}\sum_{k = 1}^{K}[y_i = k]\log c_k\right) $$

Имеем задачу условной оптимизации с условием $\sum_{k=1}^K c_k=1$. Для решения построим Лагранжиан и найдём его минимум. 

$$ L(c, \lambda) = - \frac{1}{|R|} \sum_{(x_i, y_i) \in R} \sum_{k = 1}^{K} [y_i = k] \log c_k + \lambda \sum_{k = 1}^{K} c_k \to \min_{c_k} $$

Продифференциируем

$$ \frac{\partial}{\partial c_k} L(c, \lambda) = - \frac{1}{|R|} \sum_{(x_i, y_i) \in R} [y_i = k] \frac{1}{c_k} + \lambda = - \frac{p_k}{c_k} + \lambda = 0 $$

$c_k = \frac{p_k}{\lambda}$ суммируем по $k$

$$  1 = \sum_{k = 1}^{K} c_k = \frac{1}{\lambda} \sum_{k = 1}^{K} p_k = \frac{1}{\lambda}$$

Отсюда $\lambda = 1$, Значит, минимум достигается при $c_k = p_k$.
Подставляя эти выражения в критерий, получим:

$$ H(R) = - \sum_{k = 1}^{K} p_k \log p_k$$

```python
from sklearn.datasets import load_boston
data = load_boston()

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
scoring='r2'

max_features=[None, 'log2', 'sqrt']

max_depth =[2, 4, 6, 8, 10, 20, 50]

min_samples_leaf=np.arange(1,20)
```

```python
X = data.data
y = data.target

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.3)
sp = KFold(3,shuffle =True)
```

```python
gs = GridSearchCV(DecisionTreeRegressor(random_state = 23),
                  param_grid={
                      'max_features': max_features, # the number of features to consider when looking for the best split
                      'max_depth': max_depth,
                      'min_samples_leaf': min_samples_leaf,
                  },
                  cv=sp,
                  scoring=scoring)
gs.fit(X, y)
```
